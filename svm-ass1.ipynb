Q3. What is the kernel trick in SVM?

The kernel trick in Support Vector Machines (SVM) is a powerful technique that allows SVMs to perform non-linear classification by implicitly mapping input data into a higher-dimensional space without having to compute the transformation explicitly. This is done using a kernel function, which computes the dot product of the data in the higher-dimensional space directly from the input data in the original space. Here's a more detailed explanation:

### Motivation
For some datasets, a linear decision boundary (hyperplane) is insufficient to separate the classes effectively. By transforming the input features into a higher-dimensional space, it is often possible to find a hyperplane that can separate the classes. However, computing this transformation explicitly can be computationally expensive and impractical.

## The Kernel Trick
The kernel trick leverages kernel functions to avoid the explicit computation of the coordinates of the data in the higher-dimensional space. Instead, it computes the inner products between the images of all pairs of data in this high-dimensional space directly.

## Common Kernel Functions

1. Linear Kernel: For linearly separable data.
    
       K(xi ,xj )=xi*xj

2. Polynomial Kernel: For polynomial decision boundaries.

       K(xi ,xj ) = (xi*xj +c)^d
 
       where c is a constant and d is the degree of the polynomial.
   
3. Radial Basis Function (RBF) or Gaussian Kernel: For complex, non-linear decision boundaries.
      
      
        K(xi ,xj )= exp (-(||xi-xj||^2)/2ùúé^2)

       where œÉ is a parameter that controls the width of the Gaussian.

4. Sigmoid Kernel: Similar to neural networks.
  
       K(xi ,xj ) = tanh(ùõºxi * xj + C)

       where ùõº and c are kernel parameters.
       
## Benefits of the Kernel Trick
Efficiency: It allows SVMs to operate in very high-dimensional spaces efficiently.
Flexibility: By choosing different kernel functions, SVMs can model a wide variety of decision boundaries.
Non-linear Boundaries: It enables the creation of non-linear decision boundaries in the original input space.

Q4. What is the role of support vectors in SVM Explain with example

Role of Support Vectors in SVM
Support vectors are the critical elements of the training dataset in Support Vector Machines (SVM). They are the data points that lie closest to the decision boundary (or hyperplane) and are pivotal in defining the position and orientation of the hyperplane. The key role of support vectors includes:

Determining the Margin: Support vectors are the points that lie on the edges of the margin, which is the region between the hyperplanes that separate the two classes. The distance between these support vectors and the hyperplane is maximized to achieve the best separation of the classes.
Influencing the Decision Boundary: Only the support vectors contribute to the decision boundary. Points that are not support vectors do not affect the position of the hyperplane.
Model Complexity: The number of support vectors directly influences the complexity of the SVM model. A smaller number of support vectors generally leads to a simpler model.

SVM?

Illustrations and Examples of Hyperplane, Marginal Plane, Soft Margin, and Hard Margin in SVM

1. Hyperplane
The hyperplane is the decision boundary that separates the classes in SVM. In a 2D space, it is a line, and in a 3D space, it is a plane. For higher dimensions, it is a hyperplane.

The hyperplane (decision boundary) separates the two classes (represented by 'o' for one class and 'x' for the other class).

2. Marginal Planes
Marginal planes are the boundaries on either side of the hyperplane, which are parallel to it and at the maximum distance that does not enclose any data points between them in the hard margin case.

The marginal planes are shown by the dashed lines parallel to the hyperplane, creating a margin where no data points (for the hard margin case) reside.

3. Hard Margin SVM
Hard margin SVM assumes that the data is linearly separable and finds the hyperplane that maximizes the margin with no misclassification allowed.

The solid line is the hyperplane, and the dashed lines are the marginal planes. All points lie outside or on the margin, with no misclassification.

4. Soft Margin SVM
Soft margin SVM allows some misclassifications and introduces slack variables to handle non-linearly separable data. It finds a balance between maximizing the margin and minimizing the classification error.

The solid line is the hyperplane, and the dashed lines are the marginal planes. Some points (e.g., the x at (2,2)) are within the margin or misclassified, which is handled by introducing slack variables.

```python
# Load the Iris Dataset and Split into Training and Testing Sets

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
```

```python
y=iris.target
```

```python
X= iris.data
```

```python
# For simplicity, we'll only use the first two classes and the first two features
X = X[y != 2, :2]
y = y[y != 2]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test = train_test_split(X,y,test_size= 0.3 , random_state=42)
```

```python
# Train a Linear SVM Classifier Using Scikit-Learn

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

clf = SVC(C =1.0 ,kernel = "linear" )
clf.fit(X_train,Y_train)

# Predict the labels for the testing set
Y_pred = clf.predict(X_test)

# Compute the accuracy of the model
accuracy = accuracy_score(Y_test,Y_pred)
```

```python
accuracy
```

```python
import matplotlib.pyplot as plt

def plot_decision_boundaries(X, y, clf):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.coolwarm)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Decision Boundaries')
    plt.show()

plot_decision_boundaries(X_test, Y_test, clf)
```

```python

#### Try Different Values of the Regularisation Parameter \( C \)

for C in [0.01, 0.1, 1, 10, 100]:
    clf = SVC(kernel='linear', C=C)
    clf.fit(X_train, Y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"Accuracy with C={C}: {accuracy}")
    plot_decision_boundaries(X_test, Y_test, clf)
```
